## Define df_create

```python
def df_create(data, key_name):
    df = pd.DataFrame()
    for player in data['players']:
        try:
            df_concat = pd.DataFrame.from_dict(player[key_name])
        except ValueError:
            df_concat = pd.DataFrame.from_dict([player[key_name]])
        df_concat['pid'] = player['pid']
        df = pd.concat([df, df_concat])  
    return df
```

## df_draft

```python
df_draft = pd.DataFrame()

for player in data['players']:
    df_concat = pd.DataFrame.from_dict([player['draft']])
    df_concat['pid'] = player['pid']
    df_draft = pd.concat([df_draft, df_concat])
```

```python
df_draft.to_csv('draft.csv', index=False)
```

## df_ratings

```python
df_ratings = pd.DataFrame()

for player in data['players']:
    df_concat = pd.DataFrame.from_dict(player['ratings'])
    df_concat['pid'] = player['pid']
    df_ratings = pd.concat([df_ratings, df_concat])
```

```python
df_ratings.to_csv('ratings.csv', index=False)
```

## df_stats

```python
df_stats = pd.DataFrame()

for player in data['players']:
    df_concat = pd.DataFrame.from_dict(player['stats'])
    df_concat['pid'] = player['pid']
    df_stats = pd.concat([df_stats, df_concat])    
```

```python
df_stats.to_csv('stats.csv', index=False)
```
